```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
applications = pd.read_csv("C:/Users/LENOVO/Desktop/OpenClassrooms/Projet 7/data/application_train.csv")
```

```python
applications
```

```python
applications = applications[["AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_GOODS_PRICE", "AMT_ANNUITY"]]
```

```python
applications
```

```python
data = applications[
    (applications['AMT_INCOME_TOTAL'] < 550000) &
    (applications['AMT_CREDIT'] < 400000) &
    (applications['AMT_GOODS_PRICE'] < 400000) &
    (applications['AMT_ANNUITY'] < 150000)]
```

```python
data
```

```python
amt_income_total = 270000
amt_credit = 130000
amt_goods_price = 267000
amt_annuity = 27867
```

```python
graphs = data.sample(500)
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(graphs['AMT_INCOME_TOTAL'], fill=True, color="blue")
plt.axvline(x=amt_income_total, color='red', linestyle='--', linewidth=1.5)
plt.title('Total Income')
plt.xlabel('')
plt.ylabel('')  
plt.gca().axes.get_yaxis().set_visible(False)
plt.xticks(fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(graphs['AMT_CREDIT'], fill=True, color="blue")
plt.axvline(x=amt_credit, color='red', linestyle='--', linewidth=1.5)
plt.title('Amount of credit asked')
plt.xlabel('')
plt.ylabel('')  
plt.gca().axes.get_yaxis().set_visible(False)
plt.xticks(fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(graphs['AMT_GOODS_PRICE'], fill=True, color="blue")
plt.axvline(x=amt_goods_price, color='red', linestyle='--', linewidth=1.5)
plt.title('Price of goods (if applicable)')
plt.xlabel('')
plt.ylabel('')  
plt.gca().axes.get_yaxis().set_visible(False)
plt.xticks(fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(graphs['AMT_ANNUITY'], fill=True, color="blue")
plt.axvline(x=amt_annuity, color='red', linestyle='--', linewidth=1.5)
plt.title('Loan annuity')
plt.xlabel('')
plt.ylabel('')  
plt.gca().axes.get_yaxis().set_visible(False)
plt.xticks(fontsize=8)
plt.show()
```
